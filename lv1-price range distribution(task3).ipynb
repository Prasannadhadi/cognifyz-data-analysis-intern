```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Data.csv')
```

```python
df
```

```python
plt.figure(figsize=(8,6))
df['Price range'].value_counts().sort_index().plot(kind='bar',color='orange')
plt.xlabel('Price range')
plt.ylabel('Number of Restaurants')
plt.title('Distribution of price ranges among Restaurants')
plt.xticks(rotation=0)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

```python
total_restaurants=len(df)
```

```python
price_range_counts=df['Price range'].value_counts()
```

```python
percentage_restaurants=(price_range_counts/total_restaurants)*100
print('percentage restaurants in each pric range category:')
print(percentage_restaurants)
```

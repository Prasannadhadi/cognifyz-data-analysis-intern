```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('Data.csv')
```

```python
df
```

```python
def analyze_rating_distribution(dataset_path):
    plt.figure(figsize=(8,6))
    plt.hist(df['Aggregate rating'],bins=10,edgecolor='black',color='skyblue')
    plt.xlabel('rating')
    plt.ylabel('frequency')
    plt.title('Distribution of ratings')  
    plt.grid(True)
    plt.show()
             
    most_common_range=df['Aggregate rating'].mode()[0]
    return most_common_range
dataset_path='restaurants.csv'
most_common_rating_range=analyze_rating_distribution(dataset_path)
print(f' The most common rating range is:{most_common_rating_range}')
```

calculate the average number of votes received by restaurants

```python
def calculate_average_Votes(dataset_path):
    average_Votes=df['Votes'].mean()
    return average_Votes
dataset_path='restaurants.csv'
average_Votes=calculate_average_Votes(dataset_path)
print(f'The average number of Votes received by restaurants is:{average_Votes:.2f}%')
```
